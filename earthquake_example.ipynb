# Python Visualization Example: Earthquake Data

## Introduction

- Why earthquake data analysis?
  - Earthquakes can have devastating effects on communities, making it crucial to understand their patterns and impacts.
  - Analyzing earthquake data helps in disaster preparedness and response planning.


- Set-up:
  - We have earthquake data containing information on earthquake events, including their magnitudes and locations.
  - We concluded from analysis that earthquake activity is more frequent and severe in certain regions.

- Goal:
  - To visualize earthquake data to tell the story.

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# We are using some random data to simulate earthquake magnitudes
# So we set parameters and generate random samples in this cell
mean1, sd1 = 5, 2
mean2, sd2 = 10, 1.5
mean3, sd3 = 15, 0.2

mean1, sd1 = 6.69, 0.83
mean2, sd2 = 8.31, 1.25
mean3, sd3 = 13, 0.1

batch_size = 200  # number of samples to generate
# create some Gaussian distributions that might describe how people would rate a cute dog
magnitudes = np.linspace(0,20,200)

occurences_1 = np.exp(-0.5*((magnitudes-mean1)/sd1)**2)/(sd1*np.sqrt(2*np.pi))
occurences_2 = np.exp(-0.5*((magnitudes-mean2)/sd2)**2)/(sd2*np.sqrt(2*np.pi))
occurences_3 = np.exp(-0.5*((magnitudes-mean3)/sd3)**2)/(3.7*sd3*np.sqrt(2*np.pi))

loc1_magnitudes = np.random.normal(loc=mean1, scale=sd1, size=batch_size)
loc2_magnitudes = np.random.normal(loc=mean2, scale=sd2, size=batch_size)
loc3_magnitudes = np.random.normal(loc=mean3, scale=sd3, size=batch_size)
```

```python
# Example of plot generate from an equation
fig, ax = plt.subplots(figsize=(8,4))
fig.suptitle(f"Earthquake Scale")

ax.plot(magnitudes, occurences_1, linewidth=3, color='gray')
ax.plot(magnitudes, occurences_2, linewidth=3, color='magenta')
ax.plot(magnitudes, occurences_3, linewidth=3, color='dodgerblue')
```

We can visualize distributions in various ways. Some considerations include:
- Seeing outliers and their impact on the overall distribution.
- Understanding the spread and central tendency of the data.
- Comparing distributions across different groups or categories.

```python
plt.figure(figsize=(6, 4))
sns.kdeplot(loc1_magnitudes, color='gray', label='Location 1')
sns.kdeplot(loc2_magnitudes, color='magenta', label='Location 2')
sns.kdeplot(loc3_magnitudes, color='dodgerblue', label='Location 3')
plt.xlabel("Magnitude")
plt.ylabel("Number of Earthquakes")
plt.title("Earthquake Magnitude Distribution")
plt.legend();
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(data=[loc1_magnitudes, loc2_magnitudes, loc3_magnitudes])
plt.ylabel("Magnitude")
plt.title("Boxplot of Earthquake Magnitudes")
plt.xticks([0, 1, 2], ['Location 1', 'Location 2', 'Location 3']);
```

```python
plt.figure(figsize=(6, 4))
sns.violinplot(data=[loc1_magnitudes, loc2_magnitudes, loc3_magnitudes])
plt.ylabel("Magnitude")
plt.title("Violin Plot of Earthquake Magnitudes")
plt.xticks([0, 1, 2], ['Location 1', 'Location 2', 'Location 3']);
```

```python
def AI_plot():
    # Enhanced final plot with multiple features
    fig, ax = plt.subplots(figsize=(14,7))
    fig.suptitle("Earthquake Scale: Enhanced Visualization", fontsize=22, fontweight='bold', color='navy')

    # Plot the distributions
    sns.histplot(loc1_magnitudes, kde=True, fill=False, alpha=0.5, stat="density", bins=20, color='gray', label='Location 1', ax=ax)
    sns.histplot(loc2_magnitudes, kde=True, fill=False, alpha=0.5, stat="density", bins=20, color='magenta', label='Location 2', ax=ax)
    sns.histplot(loc3_magnitudes, kde=True, fill=False, alpha=0.5, stat="density", bins=5, color='dodgerblue', label='Location 3', ax=ax)

    # Add grid
    ax.grid(True, linestyle='--', alpha=0.6)

    # Axis labels
    ax.set_xlabel('Magnitude', fontsize=16, fontweight='bold')
    ax.set_ylabel('Density of Events', fontsize=16, fontweight='bold')

    # Custom ticks and tick labels
    ax.set_xticks([0, 5, 10, 13, 15, 20])
    ax.set_xticklabels(["0 (No damage)", "5 (Little damage)", "10 (Some damage)", "13 (Max damage)", "15", "20"], rotation=20)
    ax.set_yticks([0, 0.5, 1, 2, 3.5])
    ax.set_yticklabels(["No events", "Some events", "Many events", "High", "Extreme"])

    # Add legend with title
    ax.legend(title='Epicenter Location', fontsize=12, title_fontsize=14, loc='upper right', frameon=True, shadow=True)

    # Add vertical lines for means
    ax.axvline(mean1, color='gray', linestyle=':', linewidth=2, label='Mean 1')
    ax.axvline(mean2, color='magenta', linestyle=':', linewidth=2, label='Mean 2')
    ax.axvline(mean3, color='dodgerblue', linestyle=':', linewidth=2, label='Mean 3')

    # Add shaded regions for damage zones
    ax.axvspan(0, 5, color='green', alpha=0.08, label='Low Damage Zone')
    ax.axvspan(5, 10, color='yellow', alpha=0.08, label='Moderate Damage Zone')
    ax.axvspan(10, 20, color='red', alpha=0.05, label='High Damage Zone')

    # Add annotation arrows and text
    annotations = [
        (mean1, 0.5, 'Location 1', 'gray'),
        (mean2, 0.22, 'Location 2', 'magenta'),
        (mean3, 3.3, 'Location 3', 'dodgerblue'),
    ]
    text_positions = [(mean1-5, 0.8), (mean2-2, 0.5), (mean3-6, 3.0)]
    for (point_x, point_y, text, color), text_pos in zip(annotations, text_positions):
        ax.annotate(
            text.upper(),
            xy=(point_x, point_y),
            xytext=text_pos,
            size=16,
            color=color,
            arrowprops=dict(arrowstyle='fancy', color=color, lw=2, alpha=0.7),
            bbox=dict(boxstyle='round,pad=0.3', fc='white', ec=color, lw=1.5, alpha=0.7)
        )

    # Add text box for explanation
    ax.text(0.98, 0.02,
            'Shaded regions indicate damage zones.\nDashed lines show mean magnitude per location.',
            transform=ax.transAxes, fontsize=12, color='navy',
            ha='right', va='bottom', bbox=dict(boxstyle='round', facecolor='white', alpha=0.7, edgecolor='navy'))

    # Set background color
    ax.set_facecolor('#f7f7fa')
    fig.patch.set_facecolor('#eaf0fa')

    # Tight layout for better spacing
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])

    # Show the plot
    plt.show()
```

```python
AI_plot()
```
